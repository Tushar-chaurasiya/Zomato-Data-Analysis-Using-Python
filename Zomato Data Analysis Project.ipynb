# Cleaning Rate Column

```python
# Get unique values in the 'rate' column
df['rate'].unique()
```

# Removing "NEW", "-" and "/5" from Rate Column

```python
# Define a function to handle 'rate' column values
def handlerate(value):
    # Check if the value is 'NEW' or '-' and replace it with NaN
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        # Split the string at '/' and take the first part as the rating
        value = str(value).split('/')
        value = value[0]
        return float(value)   # Convert the rating to a float

# Apply the handlerate function to the 'rate' column
df['rate'] = df['rate'].apply(handlerate)
# Display the first 5 values of the 'rate' column to check the result
df['rate'].head()
```

```python
# Count the number of missing (NaN) values in the 'rate' column
df.rate.isnull().sum()
```

# Filling Null Values in Rate Column with Mean

```python
# Fill missing values in the 'rate' column with the mean of the 'rate' column
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()    # Check if there are any missing values left in the 'rate' column
```

```python
# Get a concise summary of the DataFrame, including data types and non-null values
df.info()  
```

# Dropping Null Values

```python
# Drop any rows with missing values (NaN) from the DataFrame
df.dropna(inplace = True)
df.head()    # Display the first five rows of the updated DataFrame
```

```python
# Rename the columns: 'approx_cost(for two people)' to 'Cost2plates' 
# and 'listed_in(type)' to 'Type' in the DataFrame
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()    # Display the first five rows of the updated DataFrame
```

```python
# Display the first five rows of the updated DataFrame after dropping the 'listed_in(city)' column
df.head()
```

```python
# Get the unique values from the 'Cost2plates' column
df['Cost2plates'].unique()
```

# Removing( , ) from Cost2Plates Column

```python
# Define a function to handle commas in numeric values
def handlecomma(value):
    value = str(value)    # Convert the value to a string
    if ',' in value:      # Check if there's a comma in the string
        value = value.replace(',', '')    # Remove the comma
        return float(value)     # Convert the cleaned string to a float
    else:
        return float(value)     # Convert directly to float if no comma

# Apply the handlecomma function to the 'Cost2plates' column to clean the values
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()     # Get the unique values from the cleaned 'Cost2plates' column
```

```python
# Display the first five rows of the DataFrame to check the updated 'Cost2plates' values
df.head()
```

# Cleaning Rest Type Column

```python
# Get the count of unique values in the 'rest_type' column
df['rest_type'].value_counts()
```

```python
# Get the count of unique values in the 'rest_type' column, sorted in descending order
rest_types = df['rest_type'].value_counts(ascending = False)

# Display the counts of each restaurant type
rest_types     
```

```python
# Filter the restaurant types that have counts less than 1000
rest_types_lessthan1000 = rest_types[rest_types < 1000]

# Display the filtered restaurant types with counts less than 1000
rest_types_lessthan1000
```

# Making Rest Types less than 1000 in frequency as others

```python
# Define a function to categorize less common restaurant types as 'others'
def handle_rest_type(value):
    # Check if the restaurant type is in the less common types
    if(value in rest_types_lessthan1000):
        return 'others'     # Replace with 'others' if it's less common
    else:
        return value        # Return the original value if it's common
    
# Apply the handle_rest_type function to the 'rest_type' column to update the values
df['rest_type'] = df["rest_type"].apply(handle_rest_type)

# Get the count of unique values in the updated 'rest_type' column
df['rest_type'].value_counts()
```

```python
# Display the first five rows of the updated DataFrame to check changes in the 'rest_type' column
df.head()
```

# Cleaning Location Column

```python
# Get the count of unique values in the 'location' column
df['location'].value_counts()
```

```python
# Get the count of unique values in the 'location' column, sorted in descending order
location = df['location'].value_counts(ascending = False)

# Filter the locations that have counts less than 300
location_lessthan300 = location[location < 300]

# Define a function to categorize less common locations as 'others'
def handle_location(value):
    # Check if the location is in the less common types
    if(value in location_lessthan300):
        return 'others'     # Replace with 'others' if it's less common
    else:
        return value      # Return the original value if it's common

# Apply the handle_location function to the 'location' column to update the values
df['location'] = df['location'].apply(handle_location)

# Get the count of unique values in the updated 'location' column
df['location'].value_counts()
```

```python
# Display the first five rows of the updated DataFrame to check changes in the 'location' column
df.head()
```

# Cleaning Cuisines Column

```python
# Get the count of unique values in the 'cuisines' column, sorted in descending order
cuisions = df['cuisines'].value_counts(ascending = False)

# Filter the cuisines that have counts less than 100
cuisions_lessthan100 = cuisions[cuisions < 100]

# Define a function to categorize less common cuisines as 'others'
def handle_cuisions(value):
    # Check if the cuisine is in the less common types
    if(value in cuisions_lessthan100):
        return 'others'     # Replace with 'others' if it's less common
    else:
        return value     # Return the original value if it's common
    
# Apply the handle_cuisions function to the 'cuisines' column to update the values    
df['cuisines'] = df['cuisines'].apply(handle_cuisions)

# Get the count of unique values in the updated 'cuisines' column
df['cuisines'].value_counts()
```

```python
# Display the first five rows of the updated DataFrame to check changes in the 'cuisines' column
df.head()
```

# No. of Votes, location Wise

```python
# Select 'location' and 'votes' columns and create a new DataFrame df4
df4 = df[['location','votes']]

# Drop duplicate rows in df4
df4.drop_duplicates()

# Group by 'location' and sum the 'votes' for each location
df5 = df4.groupby(['location'])['votes'].sum()

# Convert the resulting Series to a DataFrame
df5 = df5.to_frame()

# Sort the DataFrame by 'votes' in descending order
df5 = df5.sort_values('votes', ascending= False)

# Display the top 10 locations by votes
df5.head(10)
```

```python
plt.figure(figsize= (15,8))     # Set the figure size for the plot
sns.barplot(x= df5.index, y= df5['votes'])     # Create a bar plot for the total votes by location
plt.xticks(rotation=90)     # Rotate x-axis labels by 90 degrees for better readability
```
